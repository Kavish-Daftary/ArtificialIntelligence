```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

```
dataset = pd.read_csv('/content/datasets_13720_18513_insurance.csv')
dataset
```

```

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

gender=encoder.fit_transform(dataset['sex'])
smoker=encoder.fit_transform(dataset['smoker'])
region=encoder.fit_transform(dataset['region'])
```

```
print(gender)
print(smoker)
print(region)
```

```
dataset.iloc[:,1]=gender
dataset.iloc[:,4]=smoker
dataset.iloc[:,5]=region
```

```
dataset
```

```
X = dataset.iloc[:,0:6].values
y = dataset.iloc[:,6].values
```

```
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers  import SGD, Adam

model = Sequential()

model.add(Dense(6, input_shape=(6,), activation='tanh'))
model.add(Dense(10, activation='softmax'))
model.add(Dense(10, activation='softmax'))
model.add(Dense(1, activation='softmax'))

model.compile(Adam(lr=0.04), 'categorical_crossentropy', metrics=['accuracy'])

model.summary()
```
